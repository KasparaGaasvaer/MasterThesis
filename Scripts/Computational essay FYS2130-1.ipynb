## Hva er akustisk impedans?

### Refleksjon og transmisjon av bølger 

Der vi opplever et skifte mellom ulike medier av ulik akustisk impedans, kaller vi gjerne et grensesjikte. Her vil vi kunne se refleksjon eller transmisjon, men stor sett begge deler samtidig. Den bølgen som er på vei mot grensesjiktet, men ikke enda har nådd det kaller vi *hendelsesbølgen* eller den inkommende bølgen. Den *reflekterte* bølgen er bølgen som beveger seg vekk fra grensesjiktet, men fremdeles i samme mediet som den inkommende bølgen. Bølgen som beveger seg vekk fra grensesjiktet, men da i det nye mediet på motsatt siden av der den reflekterte bølgen beveger seg, kaller vi den *transmitterte* bølgen. 





### Akustisk impedans 

Begrepet akustisk impedans betegner det vi gjerne kaller ledningsevnen til lydbølger. Denne avhenger av den hastigheten til bølgen som sendes gjennom et mediet, $c$, og massetettheten til dette mediet, $\rho$. Hvert vevesmediet har en karakteristisk impedans, og vi kan finne denne ved likningen $$ Z_0 = \rho c$$

Har vi en bølge som beveger seg fra et medie med impedans $z_a$ til et med impedans $z_b >> z_a$, der $z_b \rightarrow \infty$, vil bølgen bli totalt reflektert, der den reflekterte bølgen har negativ fortegn. Har det nye mediet større impedans, men ikke så stor at den går mot uendelig, vil bølgen bli delvis reflektert og delvis transmittert når den treffer grensesjiktet. Den reflekterte bølgen vil i dette tilfellet også har negativt fortegn, mens den transmitterte har likt fortegn som den inkommende bølgen. Ved tilfellet der impedansen til det initielle mediet $z_a$ er større enn det nye mediet, $z_b<z_a$, vil noe av bølgen bli reflektert, og noe transmittert. I motsetning til tidligere tilfeller vil den reflekterte bølgen nå ha likt fortegn som den inkommende, positivt. For et tilfelle der det nye mediet bølgen beveger seg mot er $z_b<<z_a$ der $z_b \rightarrow 0$, vil hele bølgen bli reflekter med positivt fortegn.

For å finne ut hvor mye av den inkommende bølgen som reflekteres og transmitteres, finner vi det vi kaller refleksjonskoeffisienten og transmisjonskoeffisienten. Vi ser da på forholdet mellom amplitudene til den inkommende bølgen $A_i$, den reflekterte bølgen $A_r$ og den transmitterte bølgen $A_t$. Refleksjonskoeffisienten R er da gitt ved $$R = \frac{A_r}{A_i} = \frac{z_a-z_b}{z_a+z_b}$$ og transmisjonskoeffisienten T ved $$T=\frac{A_t}{A_i}=\frac{2z_a}{z_a+z_b}$$ Det finnes flere enheter for akustisk impedans, men i dette essayet brukes $Pas/m$. Noen av kildene kan bruke andre enheter, f.eks. Rayl.

### Ultralyd

Per definisjon er ultralyd en mekanisk bølge som ligger på en frekvens mellom 16 kHz og 1 GHz, og i væsker og vev slik vi skal studere, består ultralyd av longitudinalbølger. Vi bruker bølger med høy frekvens og lav bølgelengde av flere grunner når vi utfører slike undersøkelser. Den viktigste av dem er at diffraksjonen blir betydelig mindre jo høyere frekvens bølgen har når den treffer en hindring, og resultatet blir da mer presist. Det å bruke bølger med samme frekvens som hørbar lyd er heller ikke optimalt når det kommer til pasientpleie, da dette anatkeligvis vil være mer urovekkende enn en behandlig som er støyløs. 

Transduseren er apparatet som brukes til å sende ultralydbølger inn i kroppen. Den vet når bølgen sendes ut, hastigheten, retningen, og når den blir registrert av transduseren igjen. For hvert grensesjiktet i kroppen bølgen treffer, altså for hvert nye medie bølgen beveger seg gjennom, blir et ekko av bølgen reflektert tilbake. Noe av den beveger seg videre til neste grensesjiktet, og reflekterer igjen tilbake ny informasjon. Fra dette kan da en datamaskin tolke resultatene og produsere et todimensjonalt bilde som avslører mye av innsiden av kroppen. 

Under er en animajson av en puls som sendes fra et punkt (transduseren), og treffer et grensesjikte. Noe reflekteres tilbake til transduseren, mens noe av følgen beveger seg videre og treffer et nytt grensesjikte.

```python
%matplotlib notebook 
from matplotlib import animation

#Definerer grenseverdier
fig = plt.figure()
y_ned = -10.0
y_opp = -y_ned
x_venstre = -10.0
x_høyre = 20.0


ax = plt.axes(xlim=(x_venstre,x_høyre),ylim=(y_ned,y_opp))
ax.set_aspect("equal")
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])

#Lager linjer 
lines = []
bølge_r1=ax.plot([],[],lw=2,color="black")[0] #første reflekterte bølge
lines.append(bølge_r1)
punkt=ax.plot([],[],linestyle="none",marker="o",markersize=3,color="black")[0] #startpunkt til bølgen
lines.append(punkt)
vegg1=ax.plot([],[],lw=2,color="red")[0] # første grensevegg
lines.append(vegg1)
bølge_r2=ax.plot([],[],lw=2,color="blue")[0] #andre reflekterte bølge
lines.append(bølge_r2) 
vegg2=ax.plot([],[],lw=2,color="green")[0] # andre grensevegg
lines.append(vegg2)

#Funksjon for relevante linjer over
def all_lines2D():
    for line in lines:
        line.set_data([],[])
    return lines

#Definerer parametere
t_min = 0; t_max = 10.0
n_theta = 200; theta = np.linspace(0,2*np.pi,n_theta)
n_time=900; time = np.linspace(t_min,t_max,n_time)
v = 1

#Funksjon som tar inn linjene og animerer 
def animate2D(k):
    r = v*time[k] 
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    x_1 = x.copy(); y_1 = y.copy()
    x_2 = x.copy(); y_2 = y.copy()
    x_1[x>=4] = -x[x>=4] + 8
    x_2[x>=8] = -x[x>=8] + 16
    
    for l_number,line in enumerate(lines):
        if l_number==0:
            line.set_data(x_1,y_1)
        if l_number==1:
            line.set_data([0],[0])
        if l_number==2:
            line.set_data([4,4],[y_ned,y_opp])
        if l_number==3:
            line.set_data(x_2,y_2)
        if l_number==4:
            line.set_data([8,8],[y_ned+3,y_opp-3])
    return lines    

n = len(time)-1
animasjon2D = animation.FuncAnimation(fig, animate2D, init_func=all_lines2D,frames=n, interval=2, blit=True)
```

### Ultralyd av nyresten uten gel

Vi ser på et av de første stegene ved en ultralyd undersøkelse, når bølgen sendes fra transduseren og inn i kroppen. I dette tilfellet vil bølgen starte i mediet luft, som har en akustisk impedans på $0.0004 MPas/m$. Grensesjiktet den treffer er da huden, og bølgen beveger seg så i en impedans på ca. $1.69 MPas/m $. Ved å bruke likningen vi har for å regne ut refleksjonskoeffisienten og transmisjonskoeffisienten, finner vi de de nye amplitudene til den reflekterne og transmitterte bølge. Under er bølgen som går fra luft til bløtvev (hud) animert.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import math

#Definerer grensverdier
fig = plt.figure()
A = 4
x_venstre  = -20.0
x_høyre = -x_venstre
y_opp = 8
y_ned = -8
ax = plt.axes(xlim=(x_venstre,x_høyre),ylim=(y_ned,y_opp))
ax.set_aspect("equal")
ax.axis("off")

#Lager linjer
lines = []
bølge_v=ax.plot([],[],lw=1,color="black")[0]  #bølge fra venstre
lines.append(bølge_v)
bølge_h=ax.plot([],[],lw=1,color="red",linestyle="dotted")[0]  #bølge fra høyre
lines.append(bølge_h)
bølge_r=ax.plot([],[],lw=1,color="black")[0]  #bølge som skal reflekteres
lines.append(bølge_r)
vegg=ax.plot([],[],lw=2,color="red")[0] #grensevegg
lines.append(vegg)
punkt=ax.plot([],[],linestyle="none",marker="o",markersize=4,color="red")[0] 
lines.append(punkt)

#Funksjon for relevante linjer 
def all_lines():
    for line in lines:
        line.set_data([],[])
    return lines

#Definerer parametere
x=np.linspace(x_venstre,x_høyre,501)
lx = int(len(x)/2)
v = 1.0 #wave speed
t_min = -20.0
t_max = 30.0
time = np.linspace(t_min,t_max,700)
f = 1 #MHz, setter frekvensen ti bølgen å være innenfor grensene til ultralyd
lmbda = 1/f

#Funksjon som tarc inn linjene og animerer 
def animate(k):
    for line_number,line in enumerate(lines):
        if k > 277:
            A1 = 0.002
            A2 = 3.96
        else:
            A1 = 4
            A2 = 4
        y1 = A1*np.exp(-((x-time[k])/lmbda)**2)
        y2 = -A2*np.exp(-((x+time[k])/lmbda)**2)
        y3 = y1 + y2
        y1[:251] = [None] * 251
        y2[:251] = [None] * 251
        if line_number==0:
            line.set_data(x,y1)
        #if line_number==1:
         #   line.set_data(x,y2-b)
        if line_number==2:
            line.set_data(x[0:lx],y3[0:lx])
        if line_number==3:
            line.set_data([0,0],[y_ned,y_opp])
        if line_number==4:
            line.set_data([0,0],[y3[lx],y3[lx]])


    return lines    

n = len(time)-1
anim = animation.FuncAnimation(fig, animate, init_func=all_lines,frames=n, interval=1)
plt.show()
```

Animasjonen i figur 2 viser oss ingen tydelig transmittert bølge, men vi ser heller at store deler av den inkommende bølgen blir reflektert tilbake. Hele $99.9\%$ blir reflektert tilbake til transduseren. Nedenfor skal vi se hvordan bølgen beveger seg videre etter at den har gått gjennom ulike medier i kroppen, treffer en nyrestein, og senden tilbake til transduseren.
